```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('young_survey.csv')
df.corr()
```

```python
brunch_df = df.corr()['Getting up']
```

```python
brunch_df[1:19]
```

```python
brunch_df[1:19].sort_values(ascending=True)
```

* 'Getting up'은 얼마나 아침에 일어나는 것을 어려워하는지에 대한 응답!
-> 값이 높을수록 늦게 일어나는 사람, 낮을수록 일찍 일어나는 사람 의미!
-> 상관계수가 +로 큰 건, 늦게 일어나는 사람이 (상대적으로) 더 좋아하는 장르,
   상관계수가 -로 큰 건, 일찍 일어나는 사람이 (상대적으로) 더 좋아하는 장르라고 할 수 있음
