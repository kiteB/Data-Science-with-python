```python
import pandas as pd
```

```python
liverpool_df = pd.read_csv('liverpool.csv', index_col=0)
```

```python
liverpool_df
```

column 이름을 앞 글자만 다 소문자로 바꿔주고 싶음

```python
liverpool_df.rename(columns={'position': 'Position'})
```

```python
liverpool_df
```

데이터 프레임을 다시 출력해보면 원래대로 나옴! rename 함수는 기존의 데이터프레임을 건들지 않으므로!

```python
liverpool_df.rename(columns={'position': 'Position'}, inplace=True)
liverpool_df
```

```python
liverpool_df
```

이제는 아예 바뀌었다는 것을 확인할 수 있음!

```python
liverpool_df.rename(columns={'position': 'Position', 'born':'Born', 'number':'Number', 'nationality':'Nationality'}, inplace=True)
```

```python
liverpool_df
```

```python
liverpool_df.index.name = 'Player Name'
```

```python
liverpool_df
```

조금 안 예쁘기는 하지만, 위와 같은 방식도 많이 나오므로 알아두기!

```python
liverpool_df.set_index('Number')
```

기존에 인덱스로 있던 선수 이름이 날아가버림! -> 인덱스를 바꾸기 전에 기존 인덱스는 새로운 column에 지정해주어야 함.

```python
liverpool_df['Player Name'] = liverpool_df.index
```

```python
liverpool_df
```

```python
liverpool_df.set_index('Number')
```

```python
liverpool_df.set_index('Number', inplace=True)
```
